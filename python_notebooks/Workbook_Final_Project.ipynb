```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import bqplot as bq
from bqplot import Tooltip
import ipywidgets as widgets
import json
import altair as alt
df = pd.read_csv("Electric_Vehicle_Population_Data.csv")
df
```

```python
dfcut = df.drop(columns=['VIN (1-10)', 'Postal Code', 'Legislative District', 'DOL Vehicle ID', 'Electric Utility', '2020 Census Tract'])
dfc= dfcut.dropna()
dfc.isna().sum()
dfcsea = dfc[dfc['City'] == 'Seattle']
```

```python
dfc['Electric Vehicle Type Short'] = dfc['Electric Vehicle Type'].replace({
    'Plug-in Hybrid Electric Vehicle (PHEV)': 'PHEV',
    'Battery Electric Vehicle (BEV)': 'BEV'
})
```

```python
dfc = dfc[dfc['Electric Range'] != 0]
dfc
```

```python
# Create a widget
make_dropdown = widgets.Dropdown(
    options=sorted(dfc['Make'].unique()),
    description='Make:',
    value=dfc['Make'].unique()[0]
)
```

```python
# Function to prepare driver plot data
def getdriverdata(make):
    counts = dfc[dfc['Make'] == make]['Electric Vehicle Type Short'].value_counts()
    return counts.index, counts.values

# driver plot
initial_x, initial_y = getdriverdata(make_dropdown.value)

x_sc_driver = bq.OrdinalScale()
y_sc_driver = bq.LinearScale()

driver_plot = bq.Bars(x=initial_x,y=initial_y,scales={'x': x_sc_driver, 'y': y_sc_driver})

x_ax_driver = bq.Axis(label='Electric Vehicle Type', scale=x_sc_driver, side='bottom')
y_ax_driver = bq.Axis(label='Count', scale=y_sc_driver, orientation='vertical')

fig_driver = bq.Figure(
    marks=[driver_plot],
    axes=[x_ax_driver, y_ax_driver],
    title='Count of Electric Vehicle Types',
    fig_margin={'top': 50, 'bottom': 180, 'left': 60, 'right': 20}
)
fig_driver
```

```python
# Function to prepare driven plot data
def getdrivendata(make):
    meanranges = dfc[dfc['Make'] == make].groupby('Model')['Electric Range'].mean()
    return meanranges.index, meanranges.values

# driven plot
initial_x_driven, initial_y_driven = getdrivendata(make_dropdown.value)
x_sc_driven = bq.OrdinalScale()
y_sc_driven = bq.LinearScale()

driven_plot = bq.Bars(x=initial_x_driven,y=initial_y_driven,scales={'x': x_sc_driven, 'y': y_sc_driven},colors=["orange"])
x_ax_driven = bq.Axis(label='Model', scale=x_sc_driven, side='bottom',label_offset='100px', tick_style={'text-anchor': 'end', 'transform': 'rotate(-90deg)'} )
y_ax_driven = bq.Axis(label='Mean Electric Range (miles)', scale=y_sc_driven, orientation='vertical')

fig_driven = bq.Figure(
    marks=[driven_plot],
    axes=[x_ax_driven, y_ax_driven],
    title='Mean Electric Range by Model',
    fig_margin={'top': 50, 'bottom': 180, 'left': 60, 'right': 20}
)
fig_driven
```

```python
# Function to update the plots when a new Make is selected
def update_plots(change):
    new_driverx, new_drivery = getdriverdata(change['new'])
    driver_plot.x = new_driverx
    driver_plot.y = new_drivery
    new_drivenx, new_driveny = getdrivendata(change['new'])
    driven_plot.x = new_drivenx
    driven_plot.y = new_driveny
```

```python
make_dropdown.observe(update_plots, names='value')
```

```python
dashboard = widgets.VBox([
    make_dropdown,
    widgets.HBox([fig_driver, fig_driven])
])

dashboard
```

```python
make_counts = dfc['Make'].value_counts()

top10_makes = make_counts.head(10).index

dfctop10 = dfc[dfc['Make'].isin(top10_makes)]

dfctop10
```

```python
make_type_counts = (
    dfctop10.groupby(['Make', 'Electric Vehicle Type Short'])
    .size()
    .reset_index(name='Count')
)

make_selection = alt.selection_single(fields=['Make'])

# Driver plot
driver_chart = alt.Chart(make_type_counts).mark_bar().encode(
    x=alt.X('Make:N', title='Make', axis=alt.Axis(labelAngle=-45,titleFontSize=15)),  
    y=alt.Y('Count:Q', title='Count', scale=alt.Scale(domain=[0, 30000]),axis=alt.Axis(titleFontSize=15)),  
    color=alt.Color('Electric Vehicle Type Short:N', legend=alt.Legend(orient='top-left')),
    tooltip=['Make', 'Electric Vehicle Type Short', 'Count']
).add_selection(
    make_selection
).properties(
    title='Top 10 Electric Vehicle Types by Make',
    width=250,  
    
    height=450
)

driven_data = (
    dfctop10.groupby(['Make', 'Model'])['Electric Range']
    .mean()
    .reset_index(name='Mean Electric Range')
)

# Driven plot
driven_chart = alt.Chart(driven_data).mark_bar(color='orange').encode(
    x=alt.X('Model:N', title='Model', axis=alt.Axis(labelAngle=-90,titleFontSize=15)),  
    y=alt.Y('Mean Electric Range:Q', title='Mean Electric Range (miles)', scale=alt.Scale(domain=[0, 300]), axis=alt.Axis(titleFontSize=15)),  
    tooltip=['Model', 'Mean Electric Range']
).transform_filter(
    make_selection
).properties(
    title='Mean Electric Range by Model',
    width=400,
    height=450
)

# Combine the charts 
dashboard = alt.hconcat(driver_chart, driven_chart).resolve_scale(y='independent').configure_title(
    fontSize=20)

dashboard
```

```python
dfh = pd.read_csv("Electric_Vehicle_Population_Size_History_By_County.csv")
dfh
```

```python
dfh['date'] = pd.to_datetime(dfh['Date'])

dfh['year'] = dfh['date'].dt.year
dfh_filtered = dfh[(dfh['year'] <= 2024) & (dfh['State'] == 'WA')]
dfh_yearly_ev = dfh_filtered.groupby('year')['Electric Vehicle (EV) Total'].sum().reset_index()

yearlyplot = alt.Chart(dfh_yearly_ev).mark_line(point=True, color="red").encode(
    x=alt.X('year:O', title='Year'),
    y=alt.Y('Electric Vehicle (EV) Total:Q', title='Total Electric Vehicles'),
    tooltip=['year:O', 'Electric Vehicle (EV) Total:Q']
).properties(
    width=500,
    height=350,
    title="Total Number of Electric Vehicles Per Year in Washington State"
)

yearlyplot
```

```python
yearlyplot.save(jekdir+"line.json")
```

```python
# Group the data by year and calculate the sum of 'Electric Vehicle (EV) Total' and 'Total Vehicles' for each year
dfh_yearly = dfh_filtered.groupby('year').agg({
    'Electric Vehicle (EV) Total': 'sum',
    'Total Vehicles': 'sum'
}).reset_index()

# Calculate the percentage of electric vehicles for each year
dfh_yearly['Percent Electric Vehicles'] = (dfh_yearly['Electric Vehicle (EV) Total'] / dfh_yearly['Total Vehicles']) * 100

# Create the Altair bar chart
bar_chart = alt.Chart(dfh_yearly).mark_bar().encode(
    x=alt.X('year:O', title='Year'),
    y=alt.Y('Percent Electric Vehicles:Q', title='Percent Electric Vehicles'),
    tooltip=['year:O', 'Percent Electric Vehicles:Q']
).properties(
    width=500,
    height=350,
    title="Electric Vehicles Percentage Per Year in Washington State"
)

bar_chart
```
